## Basic Plotting

```python
sns.boxplot(df['Fare'])
plt.title("Boxplot of Fare")
plt.xlabel("Fare")
plt.show()
```

```python
sns.histplot(df['Age'], bins=20, kde=True,color='crimson')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
numerical = df.select_dtypes(include=np.number)

sns.heatmap(numerical.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap") 
plt.show() 
```

```python
df["Age"].plot(kind='hist',bins=20,color='green', alpha=0.7,)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(axis='y',linestyle='--', alpha=0.7)
```

```python
survived_counts = df["Survived"].value_counts()
labels = ['Did not survive','survived']
plt.pie(survived_counts,labels=labels, colors=['lightcoral', 'skyblue'])
plt.title("Survival Portion")
plt.show()
```

## Univariate Analysis 

### Numerical Data 

```python
figure = (figsize=(15,5))
sns.histplot(df['Fare'], kde=True, color='blue')
plt.title("Distribution of Fare")
plt.show()
```

```python
sns.boxplot(x=df['Age']) 
plt.title("Boxplot of Age") 
plt.show() 
```

### Categorical Data

```python
sns.countplot(x='Pclass', data=df, palette='viridis') 
plt.title("Passenger Count by Class") 
plt.xlabel("Passenger Class") 
plt.ylabel("Count") 
plt.show()
```

```python
df['Embarked'].value_counts().plot(kind='bar',color='orange')
```

## Bivariate Analysis 

### Numerical-Numerical 

```python
sns.scatterplot(x='Age', y='Fare', data=df) 
plt.title("Scatterplot of Age vs Fare") 
plt.xlabel("Age") 
plt.ylabel("Fare") 
plt.show() 
```

### Categorical-Categorical 

```python
sns.countplot(x='Pclass', hue='Survived', data=df) 
plt.title("Survival Count by Passenger Class") 
plt.xlabel("Passenger Class") 
plt.ylabel("Count") 
plt.show() 
```

### Numerical - Categorical

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Boxplot of age and survival status") 
plt.xlabel("Survived") 
plt.ylabel("Age") 
plt.show() 
```

## Multivariate Analysis

```python
sns.pairplot(df[['Age','Fare','Survived','Pclass']],hue='Survived',palette='coolwarm')
plt.suptitle("Pairplot of Selected Features", y=1.02) 
plt.show() 
```
